```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
resumes_job_postings = pd.read_csv('../../data/resumes_job_postings.csv')
resumes_job_postings.head()
```

```python
plt.figure(figsize=(8, 6))
resumes_job_postings['label'].value_counts().plot(kind='bar', title='Rozkład etykiet')
plt.xlabel('Etykieta')
plt.ylabel('Liczba')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
resumes_job_postings['resume_text'].apply(len).plot(kind='hist', bins=50, title='Długość CV')
plt.xlabel('Długość resume_text')
plt.ylabel('Częstotliwość')

plt.subplot(1, 2, 2)
resumes_job_postings['job_description_text'].apply(len).plot(kind='hist', bins=50, title='Długość ofert pracy')
plt.xlabel('Długość job_description_text')
plt.ylabel('Częstotliwość')

plt.tight_layout()
plt.show()
```

```python
resumes_texts = ' '.join(resumes_job_postings['resume_text'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(resumes_texts)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
job_description_texts = ' '.join(resumes_job_postings['job_description_text'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(job_description_texts)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
