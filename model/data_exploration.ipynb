```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# 1. Distribution of labels
plt.figure(figsize=(8, 6))
df['label'].value_counts().plot(kind='bar', title='Rozkład etykiet')
plt.xlabel('Etykieta')
plt.ylabel('Liczba')
plt.show()
```

```python
# 2. Length of texts in resume_text and job_description_text
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df['resume_text'].apply(len).plot(kind='hist', bins=50, title='Długość CV')
plt.xlabel('Długość resume_text')
plt.ylabel('Częstotliwość')

plt.subplot(1, 2, 2)
df['job_description_text'].apply(len).plot(kind='hist', bins=50, title='Długość ofert pracy')
plt.xlabel('Długość job_description_text')
plt.ylabel('Częstotliwość')

plt.tight_layout()
plt.show()
```
