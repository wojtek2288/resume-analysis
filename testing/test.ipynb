```python
from nltk.tokenize import sent_tokenize

resume_sentences = sent_tokenize(resume_summary[0]['summary_text'])
job_posting_sentences = sent_tokenize(job_posting_summary[0]['summary_text'])

print(resume_sentences)
print(job_posting_sentences)
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim
import numpy as np

model = SentenceTransformer(
    "jinaai/jina-embeddings-v2-small-en",
    trust_remote_code=True
)
model.max_seq_length = 1024

similarities = {}

for job_posting_sentence in job_posting_sentences:
    max_similarity = float('-inf')
    for resume_sentence in resume_sentences:
        sentences = [job_posting_sentence, resume_sentence]
        embeddings = model.encode(sentences)
        print(embeddings[0])
        similarity = cos_sim(embeddings[0], embeddings[1])

        if similarity > max_similarity:
            similarities[job_posting_sentence] = np.squeeze(similarity[0].numpy())

overall_similarity = sum(similarities.values()) / len(job_posting_sentences)
print(overall_similarity)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def create_resume_pdf(filename):
    c = canvas.Canvas(filename, pagesize=letter)
    width, height = letter

    # Function to add text with word wrap
    def draw_wrapped_text(c, text, x, y, max_width):
        lines = []
        words = text.split()
        line = []
        for word in words:
            if c.stringWidth(' '.join(line + [word]), "Helvetica", 12) <= max_width:
                line.append(word)
            else:
                lines.append(' '.join(line))
                line = [word]
        if line:
            lines.append(' '.join(line))
        for i, line in enumerate(lines):
            c.drawString(x, y - (i * 15), line)
        return y - (len(lines) * 15)

    # Personal Information
    c.setFont("Helvetica-Bold", 20)
    c.drawString(72, height - 72, "John Smith")
    c.setFont("Helvetica", 12)
    c.drawString(72, height - 92, "456 Oak Street")
    c.drawString(72, height - 107, "Silicon Valley, CA 94025")
    c.drawString(72, height - 122, "john.smith@email.com")
    c.drawString(72, height - 137, "(555) 987-6543")

    # Objective
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, height - 162, "Objective:")
    c.setFont("Helvetica", 12)
    y_position = draw_wrapped_text(c, "Results-driven Software Developer with a passion for creating efficient and innovative software solutions. Leveraging expertise in modern programming languages, web technologies, and Agile methodologies to contribute to the dynamic development team at Tech Innovations Inc.", 72, height - 180, width - 144)

    # Education
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position - 20, "Education:")
    c.setFont("Helvetica", 12)
    y_position -= 35
    c.drawString(72, y_position, "Bachelor of Science in Computer Science")
    c.drawString(72, y_position - 15, "University of California, Berkeley")
    c.drawString(72, y_position - 30, "Graduated: May 2016")
    y_position -= 50

    # Technical Skills
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position, "Technical Skills:")
    c.setFont("Helvetica", 12)
    skills = [
        "Programming Languages: Python, Java, C#",
        "Web Technologies: HTML, CSS, JavaScript, React.js, Angular",
        "Databases: MySQL, PostgreSQL",
        "Version Control: Git",
        "APIs: RESTful API development and integration",
        "Methodologies: Agile/Scrum",
        "DevOps: Docker, Kubernetes"
    ]
    for i, skill in enumerate(skills):
        c.drawString(72, y_position - 20 - (i * 15), skill)
    y_position -= 20 + (len(skills) * 15)

    # Professional Experience
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position, "Professional Experience:")
    c.setFont("Helvetica-Bold", 12)
    y_position -= 20
    c.drawString(72, y_position, "Software Developer")
    c.setFont("Helvetica", 12)
    c.drawString(72, y_position - 15, "XYZ Tech Solutions, Silicon Valley, CA")
    c.drawString(72, y_position - 30, "September 2016 - Present")
    experiences = [
        "Spearheaded the development and implementation of new software programs using Python and Java, resulting in a 20% increase in operational efficiency.",
        "Collaborated closely with the product team to understand end-user requirements and translate them into technical specifications, leading to the successful launch of multiple software products.",
        "Conducted rigorous code reviews, testing, and debugging to ensure the delivery of high-quality, scalable software solutions.",
        "Optimized SQL queries and database design to enhance the performance of existing software systems, resulting in a 30% reduction in response time."
    ]
    y_position -= 50
    for i, experience in enumerate(experiences):
        y_position = draw_wrapped_text(c, experience, 72, y_position, width - 144)

    # Projects
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position - 20, "Projects:")
    projects = [
        ("E-commerce Platform", "Led the development of a robust e-commerce platform using React.js and Node.js, facilitating seamless online transactions and increasing customer satisfaction by 25%."),
        ("Inventory Management System", "Designed and implemented an inventory management system using Java and Spring Boot, resulting in a 40% reduction in inventory errors.")
    ]
    y_position -= 35
    c.setFont("Helvetica-Bold", 12)
    for title, description in projects:
        c.drawString(72, y_position, title)
        y_position -= 15
        c.setFont("Helvetica", 12)
        y_position = draw_wrapped_text(c, description, 72, y_position, width - 144)
        y_position -= 15
        c.setFont("Helvetica-Bold", 12)

    # Certifications
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position, "Certifications:")
    certifications = [
        "Agile Certified Practitioner (ACP)",
        "Docker Certified Associate (DCA)"
    ]
    c.setFont("Helvetica", 12)
    y_position -= 20
    for cert in certifications:
        c.drawString(72, y_position, cert)
        y_position -= 15

    # Languages
    c.setFont("Helvetica-Bold", 14)
    c.drawString(72, y_position - 10, "Languages:")
    languages = [
        "English (Native)",
        "Spanish (Proficient)"
    ]
    c.setFont("Helvetica", 12)
    y_position -= 30
    for lang in languages:
        c.drawString(72, y_position, lang)
        y_position -= 15

    c.showPage()
    c.save()

# Create PDF
create_resume_pdf("John_Smith_Resume.pdf")
```
